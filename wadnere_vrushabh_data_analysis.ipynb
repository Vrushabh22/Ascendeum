```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Data_id(in).csv')
```

```python
df.info()
```

```python
df['bidder'].unique()
```

```python
df['bidder'].replace(' ', 'unknown', inplace = True)
```

```python
df['prebid_win_count'].fillna(0)
```

```python
df_summary = df.groupby('bidder').agg(
    total_bids=('response_count','sum'),
    total_win=('prebid_win_count', 'sum')
).reset_index()
```

```python
df_summary
```

```python
df_summary['win_rate'] = (df_summary['total_win']/df_summary['total_bids'])*100
```

```python
df_summary.sort_values('win_rate', ascending = False)
```
